```python
def derivative_matrix(m):
    mat = np.zeros((m,m+1))
    for i in range(m):
        mat[i,i]= m-i
    return mat
```

```python
mat = derivative_matrix(5)
print(mat)

x = np.array([5.8,-9.2,3.1,-6.2,7.2,1])
x_dash = mat @ x
print(x_dash)
```

```python
print("t=0.01", np.polyval(x_dash,0.01))
print("t=0.45",np.polyval(x_dash,0.45))
print("t=0.63",np.polyval(x_dash,0.63))
print("t=0.81",np.polyval(x_dash,0.81))
```
